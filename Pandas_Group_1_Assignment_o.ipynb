# Only focusing on the United Kingdom, we Compared salaries for full-time, part-time, and contract employees and identify any trends

```python
# Task 6: Compare salaries for full-time, part-time, and contract employees in the United Kingdom
uk_data = df2[df2['employee_residence'] == 'United Kingdom']

# Create a bar plot to compare salaries for different employment type
sns.barplot(x='employment_type', y='salary', data=uk_data, hue='work_setting')
plt.title("Salaries for Different Employment Types in the United Kingdom")
plt.xlabel("Type of salary")
plt.ylabel("Salary")
plt.legend(title='Work Setting')
plt.show()
```
